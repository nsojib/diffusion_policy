```python
import json
import pandas as pd
import matplotlib.pyplot as plt

def plot_train_and_val_loss(file_path):
    # Load the logs from the JSON lines file
    with open(file_path, "r") as f:
        raw_lines = f.readlines()
    data = [json.loads(line) for line in raw_lines]
    
    # Convert the logs into a DataFrame
    df = pd.DataFrame(data)
    
    # Print basic information
    print("Columns in log:", df.columns)
    print("Number of log entries:", len(df))
    
    # Create a figure to plot both training and validation loss
    plt.figure(figsize=(10, 6))
    
    # Plot training loss
    plt.plot(df["global_step"], df["train_loss"], label="Train Loss", marker="o", linestyle="-")
    
    # Check if validation loss is present and plot if available
    if "val_loss" in df.columns:
        # Filter out rows without validation loss
        df_val = df.dropna(subset=["val_loss"])
        plt.plot(df_val["global_step"], df_val["val_loss"], label="Validation Loss", marker="x", linestyle="-")
    
    # Set labels and title
    plt.xlabel("Global Step")
    plt.ylabel("Loss")
    plt.title("Training and Validation Loss vs Global Step")
    plt.legend()
    plt.grid(True)
    plt.show()

# Call the function with your file name
```
