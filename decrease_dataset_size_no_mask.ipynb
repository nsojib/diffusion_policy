```python
# Output file path
output_file_path = "/home/carl_lab/data_franka/selected_o40z11t15.hdf5"

demos_ola = list(f_org_ola["data"].keys())[:40] 

# Create a new HDF5 file to store the selected demos
with h5py.File(output_file_path, 'w') as f_new:
    # Create the 'data' group in the new file
    data_group = f_new.create_group('data')
    
    # Copy and rename the selected demos from each dataset
    new_demo_index = 0  # Start the demo names from 'demo_0'
    
    # For demos from ola
    for demo_name in demos_ola:
        new_name = f"demo_{new_demo_index}"
        f_org_ola.copy(f_org_ola[f"data/{demo_name}"], data_group, new_name)
        new_demo_index += 1
    
    # For demos from tazul
    for demo_name in demos_tazul:
        new_name = f"demo_{new_demo_index}"
        f_org_tazul.copy(f_org_tazul[f"data/{demo_name}"], data_group, new_name)
        new_demo_index += 1
    
    # For demos from zahid
    for demo_name in demos_zahid:
        new_name = f"demo_{new_demo_index}"
        f_org_zahid.copy(f_org_zahid[f"data/{demo_name}"], data_group, new_name)
        new_demo_index += 1

print(f"Selected demos with new names saved to {output_file_path}")
```

```python
import h5py

# Paths for the input and output HDF5 files
input_file_path = "/home/carl_lab/data_franka/imgsd_demo/demo.hdf5"

# Specify the target demo index (inclusive)
target_demo_index = 40  # Change this to the desired target (e.g., 60 demos)

output_file_path = input_file_path.replace(".hdf5", "") + f"_sub{target_demo_index}_.hdf5"

# Open the original HDF5 file
with h5py.File(input_file_path, 'r') as f_org:
    # Get the names of all demos
    all_demos = list(f_org['data'].keys())
    
    # Select demos from 'demo_0' to 'demo_target' (inclusive)
    selected_demos = [demo_name for demo_name in all_demos if int(demo_name.split('_')[-1]) <= target_demo_index]
    
    # Create a new HDF5 file to store the selected demos
    with h5py.File(output_file_path, 'w') as f_new:
        # Create the 'data' group in the new file
        data_group = f_new.create_group('data')
        
        # Copy the selected demos
        for demo_name in selected_demos:
            f_org.copy(f_org[f"data/{demo_name}"], data_group, demo_name)
            
print(f"Selected demos saved to {output_file_path}")
```
